```python
df_clean = df.copy()
```

```python
df_clean.columns = (
    df_clean.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("/", "_")
)
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
df_clean.info()
```

Data Cleaning Summary:
The dataset contained no missing values or duplicate records.
Minimal cleaning was required, limited to column name standardization
and structural consistency checks.

```python
df_clean['gender'].value_counts().plot(kind='bar', title='Gender Distribution')
```

```python
df_clean['math_score'].plot(kind='hist', bins=10, title='Math Score Distribution')
```

```python
import matplotlib.pyplot as plt

plt.boxplot(
    [df_clean['math_score'],
     df_clean['reading_score'],
     df_clean['writing_score']],
    labels=['Math', 'Reading', 'Writing']
)

plt.title("Box Plot of Student Scores")
plt.ylabel("Score")
plt.show()
```

```python
corr = df_clean[['math_score', 'reading_score', 'writing_score']].corr()
corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Between Student Scores")
plt.show()
```

Visualization Insights:

Gender distribution appears fairly balanced.

Math scores show slightly higher variability than reading and writing.

Box plot analysis indicates no extreme outliers.

Reading and writing scores are strongly positively correlated.

Key Patterns & Insights:

Reading scores have the highest average among the three subjects.

Students who completed the test preparation course consistently scored higher across all subjects.

Female students tend to perform better in reading and writing, while math scores are comparable.

Students with standard lunch show higher average scores than those with free/reduced lunch.

Strong correlation between reading and writing suggests related skill development.

Final Conclusions

The dataset was well-structured with no missing values or duplicate records, requiring minimal preprocessing.

Reading and writing scores are strongly correlated, indicating related academic skill sets.

Math scores show higher variability compared to other subjects, suggesting uneven performance levels.

Completion of the test preparation course has a positive impact on student performance across all subjects.

Socio-economic indicators such as lunch type appear to be associated with academic outcomes.
